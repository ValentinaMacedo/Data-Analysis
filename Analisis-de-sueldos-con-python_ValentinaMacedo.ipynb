5. Printear porcentaje por seniority según años de experiencia, el mapeo es:
```
Junior: de 0 hasta 2 años.

``` 

````python
dic_WorkExperience={
"Junior": 0,
"Semi-Senior": 0,
"Senior": 0, 
}

for row in list_data:
    answer = float(row[preguntas["Años de experiencia"]])
    if answer < 2:
        dic_WorkExperience["Junior"] += ((1 / len(list_data)) * 100)
    elif 2 <= answer < 5:
        dic_WorkExperience["Semi-Senior"] += ((1 / len(list_data)) * 100)
    #Asumo que una persona no va a trabajar + de 70 años
    elif 5 <= answer < 71:
        dic_WorkExperience["Senior"] += ((1 / len(list_data)) * 100)
        
for key in dic_WorkExperience:
        print(f" - {key} {round(dic_WorkExperience[key],2)}%") 
````

7. Printear porcentajes de nivel de educación formal y estado, es decir: % educación secundaria, terciaria, universitaria, postgrado, doctorado, postdoctorado, completo, incompleto y en curso para cada uno.

````python
set_educational_level = set()

for row in list_data:
    answer1 = row[preguntas["Nivel de estudios alcanzado"]].upper()
    answer2 = row[preguntas["Estado"]].upper()
    set_educational_level.add(answer1 + " " + answer2)
    #set_educational_level.add((row["Nivel de estudios alcanzado"], row["Estado"]))
    
list_educational_level = list(set_educational_level)
list_educational_level.sort()
list_educational_level.remove(' ')

dic_educational_level = {}

for i in list_educational_level:
    number_of_cases = 0
    for row in list_data:
        answer1 = row[preguntas["Nivel de estudios alcanzado"]].upper()
        answer2 = row[preguntas["Estado"]].upper()
        if answer1 + " " + answer2 == i:
            number_of_cases += 1
    dic_educational_level.update({i:number_of_cases})

number_education_answers = 0    
for key in dic_educational_level:
    number_education_answers += dic_educational_level[key]

for key in dic_educational_level:
    dic_educational_level[key] = (dic_educational_level[key] * 100) / number_education_answers 

sorted_dic_educational_level = dict(sorted(dic_educational_level.items(), key=lambda item: item[1], reverse=True))

for key in sorted_dic_educational_level:
    print(f" - {key} {round(sorted_dic_educational_level[key],2)}%") 
````
